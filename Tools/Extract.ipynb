```python
from docx import Document

def extract_bold_italic_from_docx(docx_path: str):
    """Trích xuất văn bản in đậm và in nghiêng từ DOCX."""
    doc = Document(docx_path)
    bold_texts = []
    italic_texts = []

    for para in doc.paragraphs:
        for run in para.runs:
            text = run.text.strip()
            if text:
                if run.bold:
                    bold_texts.append(text)
                if run.italic:
                    italic_texts.append(text)

    return bold_texts, italic_texts

# Sử dụng hàm
bold_texts, italic_texts = extract_bold_italic_from_docx("E:/sample.docx")
print("Bold:", bold_texts)
print("Italic:", italic_texts)
```

```python
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextBox, LTTextLine, LTChar

def extract_bold_italic_from_pdf(pdf_path: str):
    """Trích xuất văn bản in đậm và in nghiêng từ PDF."""
    bold_texts = []
    italic_texts = []

    for page_layout in extract_pages(pdf_path):
        for element in page_layout:
            if isinstance(element, (LTTextBox, LTTextLine)):
                for text_line in element:
                    if isinstance(text_line, LTTextLine):
                        text = ""
                        is_bold = False
                        is_italic = False

                        for char in text_line:
                            if isinstance(char, LTChar):
                                fontname = char.fontname.lower()
                                text += char.get_text()
                                if "bold" in fontname:
                                    is_bold = True
                                if "italic" in fontname or "oblique" in fontname:
                                    is_italic = True

                        if is_bold:
                            bold_texts.append(text)
                        if is_italic:
                            italic_texts.append(text)

    return bold_texts, italic_texts

# Sử dụng hàm
bold_texts, italic_texts = extract_bold_italic_from_pdf("E:/sample.pdf")
print("Bold:", bold_texts)
print("Italic:", italic_texts)
```
