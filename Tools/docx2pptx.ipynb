```python
from docx import Document
from typing import List, Tuple, Optional
from lxml import etree

def get_paragraph_numbering(paragraph) -> Optional[str]:
    """
    Lấy thông tin numbering (số thứ tự hoặc bullet) của đoạn văn bằng cách đọc XML.
    """
    try:
        # Truy cập phần XML của đoạn văn
        xml_element = paragraph._element
        nsmap = {'w': 'http://schemas.openxmlformats.org/wordprocessingml/2006/main'}

        # Tìm phần <w:numPr> trong XML (chứa thông tin numbering)
        numPr = xml_element.find('.//w:numPr', namespaces=nsmap)
        if numPr is not None:
            # Lấy số thứ tự hoặc bullet (nếu có)
            numId = numPr.find('.//w:numId', namespaces=nsmap)
            if numId is not None:
                return "Numbered List"  # Đánh số
            else:
                return "Bullet List"    # Bullet
    except Exception as e:
        pass
    return None

def get_paragraph_level(paragraph) -> str:
    """
    Xác định cấp độ của đoạn văn:
    - Heading: Dựa vào tên style.
    - Bullet/Numbered List: Dựa vào numbering và thụt lề.
    - Normal: Đoạn văn thường.
    """
    style_name = paragraph.style.name
    numbering = get_paragraph_numbering(paragraph)

    if style_name.startswith("Heading"):
        return style_name

    if numbering:
        # Xác định cấp độ bullet/numbering dựa vào left_indent
        if paragraph.paragraph_format.left_indent:
            indent_level = int(paragraph.paragraph_format.left_indent.pt // 20) + 1
        else:
            indent_level = 1
        return f"{numbering} Level {indent_level}"

    return "Normal"

def read_docx_as_tuples(file_path: str) -> List[Tuple[str, str]]:
    """
    Đọc nội dung file .docx và trả về list các tuple (text, level).
    """
    doc = Document(file_path)
    results = []

    for para in doc.paragraphs:
        text = para.text.strip()
        level = get_paragraph_level(para)

        if text:
            results.append((text, level))

    return results

# Đường dẫn tới file .docx
file_path = "D:/demo.docx"
content = read_docx_as_tuples(file_path)

# In ra nội dung đã trích xuất
for text, level in content:
    print(f"({text}, {level})")
```

```python
from pptx import Presentation
from pptx.util import Inches
from typing import List, Tuple

MAX_PARAGRAPHS_PER_SLIDE = 5  # Giới hạn số đoạn trên mỗi slide

def create_ppt_from_content(content: List[Tuple[str, str]], output_path: str):
    """
    Tạo file PowerPoint từ danh sách nội dung (text, level) với layout cơ bản.
    Tự động chia slide nếu nội dung quá dài.
    """
    prs = Presentation()
    slide = None
    para_count = 0

    for text, level in content:
        # Nếu hết chỗ hoặc chưa có slide -> Tạo slide mới
        if slide is None or para_count >= MAX_PARAGRAPHS_PER_SLIDE:
            slide = prs.slides.add_slide(prs.slide_layouts[1])  # Title and Content
            slide.shapes.title.text = "Nội dung chi tiết"
            para_count = 0

        content_box = slide.shapes.placeholders[1]
        p = content_box.text_frame.add_paragraph()
        p.text = text

        # Điều chỉnh mức thụt lề theo cấp độ
        if "Level" in level:
            indent_level = int(level.split()[-1]) - 1
            p.level = indent_level if indent_level >= 0 else 0

        para_count += 1

    # Lưu file PowerPoint
    prs.save(output_path)
    print(f"File PowerPoint đã được tạo: {output_path}")

# Ví dụ nội dung dài
content = [
    ("Giới thiệu", "Heading 1"),
    ("Mục tiêu nghiên cứu", "Numbered List Level 1"),
    ("Phương pháp thực hiện", "Numbered List Level 2"),
    ("Phân tích dữ liệu", "Bullet List Level 1"),
    ("Thảo luận kết quả", "Normal"),
    ("Hạn chế và khó khăn", "Normal"),
    ("Ứng dụng thực tiễn", "Bullet List Level 2"),
    ("Hướng phát triển tương lai", "Numbered List Level 1"),
    ("Tổng kết và đề xuất", "Heading 2"),
    ("Tài liệu tham khảo", "Normal")
]

# Tạo file PowerPoint
output_path = "D:/output_presentation.pptx"
create_ppt_from_content(content, output_path)
```
