```python
import matplotlib.pyplot as plt

# Xác định thanh điệu của từng từ
def get_tone(w):
    vowels = ['à', 'á', 'ả', 'ã', 'ạ',
              'ằ', 'ắ', 'ẳ', 'ẵ', 'ặ',
              'ầ', 'ấ', 'ẩ', 'ẫ', 'ậ',
              'è', 'é', 'ẻ', 'ẽ', 'ẹ',
              'ề', 'ế', 'ể', 'ễ', 'ệ',
              'ì', 'í', 'ỉ', 'ĩ', 'ị',
              'ỳ', 'ý', 'ỷ', 'ỹ', 'ỵ',
              'ò', 'ó', 'ỏ', 'õ', 'ọ',
              'ồ', 'ố', 'ổ', 'ỗ', 'ộ',
              'ờ', 'ớ', 'ở', 'ỡ', 'ợ',
              'ù', 'ú', 'ủ', 'ũ', 'ụ',
              'ừ', 'ứ', 'ử', 'ữ', 'ự'
             ]
    
    for c in w:
        if c in vowels:
            return vowels.index(c) % 5 + 1
    return 0

def get_tone_value(w):
    tone_values = [0,-1,1,-2,1,-2]
    return tone_values[get_tone(w)]
    


# Vẽ biểu đồ thanh điệu
def plot_tone_chart(sentence):
    # Tách từ trong câu
    words = sentence.split()

    # Lấy giá trị thanh điệu của từng từ
    tones = [get_tone_value(word) for word in words]

    # Vẽ biểu đồ
    plt.figure(figsize=(10, 6))
    plt.plot(range(len(words)), tones, marker='o', linestyle='-', color='skyblue', label="Thanh điệu")
    plt.xticks(range(len(words)), words, rotation=45, fontsize=12)
    plt.yticks([-2, -1, 0, 1], ["Trầm sâu", "Trầm", "Bình", "Bổng"], fontsize=12)
    plt.xlabel("Từ", fontsize=14)
    plt.ylabel("Thanh điệu", fontsize=14)
    plt.title("Biểu đồ thanh điệu của câu", fontsize=16)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.legend(fontsize=12)
    plt.tight_layout()
    plt.show()

# Ví dụ
sentence = "hôm nay trời rất đẹp"
plot_tone_chart(sentence)
```

```python
def plot_tone_charts(sentences):
    plt.figure(figsize=(12, 8))

    for i, sentence in enumerate(sentences):
        # Tách từ trong câu
        words = sentence.split()

        # Lấy giá trị thanh điệu của từng từ
        tones = [get_tone_value(word) for word in words]

        # Vẽ biểu đồ cho câu hiện tại
        plt.plot(range(len(words)), tones, marker='o', linestyle='-', label=f"Câu {i+1}: {sentence}")
        
    plt.xticks(fontsize=12)
    plt.yticks([-2, -1, 0, 1], ["Trầm sâu", "Trầm", "Bình", "Bổng"], fontsize=12)
    plt.xlabel("Từ", fontsize=14)
    plt.ylabel("Thanh điệu", fontsize=14)
    plt.title("Biểu đồ thanh điệu của các câu", fontsize=16)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.legend(fontsize=12, loc='upper left')
    plt.tight_layout()
    plt.show()

# Ví dụ
sentences = ["Mọc giữa dòng sông xanh","một bông hoa tím biếc"]
plot_tone_charts(sentences)
```
