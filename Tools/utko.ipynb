```python
def preprocess(text):
    words = text.split()
    for i in range(len(words)):
        if len(words[i]) < 3:
            words[i] = "u" + words[i]
            
    result = []
    
    for w in words:
        w = w.replace("nh", "n")
        w = w.replace("th", "t")
        w = w.replace("ph", "p")
        w = w.replace("kh", "k")
        w = w.replace("gh", "g")
        w = w.replace("ng", "g")
        w = w.replace("ngh", "g")
        result.append(w)
        
    return result
```

```python
import ctnx
```

```python
def remove_dup_chars(word):
    result = ""
    for char in word:
        if char not in result:
            result += char
    return result
```

```python
def utko_word(word):
    length = len(word)
    if length % 2 == 0:
        half_length = length // 2
        first_half = word[:half_length][::-1]
        second_half = word[half_length:][::-1]
        
        return ctnx.remove_diacritics(remove_dup_chars(first_half+second_half)).replace("q","k")
    
    middle_index = length // 2
    middle_char = word[middle_index]
    first_half = word[:middle_index][::-1]
    second_half = word[middle_index+1:][::-1]
    
    return utko_word(first_half+middle_char*2+second_half)
```

```python
def normalize(word):
    word = remove_dup_chars(word)
    if word[-1] in {"d","đ","s","h"}:
        word = word[:len(word)-1] + "t"
    return word
```

```python
def utko(text, depth):
    words = preprocess(text)
    if depth == 1:
        return "".join([normalize(utko_word(w)) for w in words])


    return utko("".join([normalize(utko_word(w)) for w in words]), depth - 1) + utko("".join([normalize(utko_word(w[::-1])) for w in words]), depth - 1) + utko(utko("".join([normalize(utko_word(w)) for w in words]), depth - 1) + utko("".join([normalize(utko_word(w[::-1])) for w in words]), depth - 1),depth-1)
```
