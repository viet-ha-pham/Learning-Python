````python
import google.generativeai as genai
import json

# Thay thế bằng API key của bạn
genai.configure(api_key='')

# Chọn mô hình
model = genai.GenerativeModel('gemini-2.0-flash')


def ask_model(prompt, context=None):
    """
    Gửi prompt đến mô hình và nhận kết quả thực tế từ Gemini API.
    """
    if context:
        prompt = context + "\n" + prompt

    response = model.generate_content(prompt)
    return response.text.strip()


# Văn bản mẫu
text = """
NHỮNG MỐI QUAN HỆ THÂN MẬT NHƯNG HÚT CẠN NĂNG LƯỢNG
Có những mối quan hệ thân thiết, tưởng như là chốn an yên, nhưng càng gắn bó lâu ta càng cảm thấy kiệt quệ. Đó là những mối quan hệ lấy đi nhiều hơn là bồi đắp, khiến ta dần đánh mất năng lượng sống.
1. Mối quan hệ mất cân bằng cảm xúc:
- Đơn phương cảm xúc: Ta luôn là người lắng nghe, an ủi, nhưng khi cần giãi bày thì người kia lại thờ ơ hoặc né tránh. Ta trở thành chỗ dựa vững chắc, nhưng không có ai để tựa vào lúc mỏi mệt.
- Không hồi đáp: Ta dành rất nhiều thời gian, tình cảm, nhưng chẳng bao giờ nhận lại sự quan tâm tương xứng. Như thể ta cứ đổ nước mãi vào một chiếc ly đã thủng đáy, đến lúc nhìn lại thì chính mình đã cạn khô.
2. Mối quan hệ đè nặng tinh thần:
- Người luôn tiêu cực: Mỗi lần gặp gỡ là một lần ta bị cuốn vào vòng xoáy than phiền, trách móc cuộc đời. Dù ta cố kéo họ ra khỏi bóng tối, họ vẫn bám chặt lấy tiêu cực và vô tình kéo ta xuống theo.
- Dựa dẫm quá mức: Khi ai đó phụ thuộc hoàn toàn vào ta về cảm xúc, quyết định, thậm chí là cả giá trị bản thân. Ta phải luôn trong trạng thái "cứu viện", dần dần cạn kiệt vì chẳng còn thời gian chăm sóc chính mình.
3. Mối quan hệ tước đoạt tự do:
- Kiểm soát dưới lớp vỏ quan tâm: Người kia luôn muốn biết ta đang làm gì, ở đâu, với ai - nhân danh sự lo lắng. Nhưng thực chất, đó là sự kiểm soát ngầm, dần bóp nghẹt tự do và làm ta nặng nề mỗi khi gần họ.
- Ganh tị hoặc cạnh tranh: Thay vì chia sẻ niềm vui khi ta thành công, họ âm thầm so sánh, rồi tìm cách làm ta thấy bất an về chính mình. Ở bên những người như vậy, ta dễ mất đi sự tự tin và cảm giác yên bình vốn có.
4. Mối quan hệ lửng lơ, không rõ ràng:
- Mập mờ cảm xúc: Ta và người ấy thân thiết hơn bạn bè, nhưng lại chẳng đủ sâu đậm để là tình yêu. Sự mập mờ kéo dài khiến ta lửng lơ, không thể tiến, cũng chẳng dám lùi, mãi mắc kẹt trong một vòng lặp mệt mỏi.
- Bạn bè "trên danh nghĩa": Hai người từng thân thiết, nhưng giờ chỉ duy trì mối liên hệ hời hợt vì thói quen hoặc xã giao. Không còn sự gắn kết thật sự, nhưng cũng không ai chủ động dứt khoát.
- Tái hợp rồi lại xa cách: Cứ mỗi lần định buông bỏ, người kia lại quay lại với những lời hứa hẹn hoặc dấu hiệu thân mật, khiến ta bối rối và mắc kẹt trong chu kỳ hợp – tan không hồi kết.
- Mối quan hệ vì lợi ích: Ở bên nhau chỉ vì một sự trao đổi ngầm - tình cảm, vật chất, địa vị -  nhưng lại cố giữ vỏ bọc thân mật. Dần dần, ta cảm thấy trống rỗng vì không thực sự được trân quý.
Khi nhận ra mình đang mắc kẹt thì sao?
Quan trọng nhất là lắng nghe bản thân, dám đặt ra ranh giới và chấp nhận rời đi khi cần. Thân mật không có nghĩa là phải đánh đổi sức khỏe tinh thần. Một mối quan hệ lành mạnh sẽ tiếp thêm năng lượng, chứ không rút cạn nó.
Và đôi khi, ta không cần phải cắt đứt hay làm rõ mọi chuyện. Chỉ cần lặng lẽ giảm tương tác, từng bước rời xa, để lại phía sau những điều khiến trái tim nặng nề. Bởi bảo vệ bản thân cũng là một dạng yêu thương - yêu lấy chính mình, để tâm hồn được an yên thêm lần nữa.
Quan trọng hơn, không phải ai cũng xấu xa hay cố ý làm tổn thương ta. Nhiều người thực sự tốt bụng, thân thiện, nhưng điều đó không có nghĩa là ta và họ sẽ luôn hợp nhau. Có những mảnh ghép đẹp, nhưng lại không vừa vặn với trái tim ta. Học cách buông tay cũng là cách để cả hai được tự do tìm thấy những kết nối phù hợp 
"""

# Bước 1: Xác định nội dung chính
step1_prompt = f"Đọc đoạn văn sau và xác định nội dung chính:\n{text}"
result1 = ask_model(step1_prompt)
print("\n🔸 Nội dung chính:\n", result1)

# Bước 2: Chia đề mục chính
depth_context = result1
step2_prompt = "Dựa trên nội dung đã tóm tắt, chia văn bản thành các đề mục chính."
result2 = ask_model(step2_prompt, context=depth_context)
print("\n🔸 Đề mục chính:\n", result2)

# Bước 3: Tóm tắt từng phần
depth_context += "\n" + result2
step3_prompt = "Với mỗi đề mục chính, phân tích chi tiết nội dung và tóm tắt thành bullet points."
result3 = ask_model(step3_prompt, context=depth_context)
print("\n🔸 Tóm tắt từng phần:\n", result3)

# Bước 4: Đặt tiêu đề
step4_prompt = "Dựa trên các phân tích, hãy đặt tiêu đề phù hợp cho văn bản." 
result4 = ask_model(step4_prompt, context=depth_context)
print("\n🔸 Tiêu đề đề xuất:\n", result4)

# Bước 5: Kiểm tra và tối ưu hóa
step5_prompt = "Kiểm tra lại cấu trúc và đề xuất cải tiến nếu cần."
result5 = ask_model(step5_prompt, context=depth_context)
print("\n🔸 Kiểm tra và tối ưu hóa:\n", result5)
````
