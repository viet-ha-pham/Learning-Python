```python
import random
def generate_sequence(n, low, high, intervals, random_state=0):
    sequence = []
    
    while len(sequence) < n:
        if sequence:
            choices = [ x for x in intervals if low <= sequence[-1] + x <= high]
            delta = random.choice(choices)
            new_item = sequence[-1] + delta
            if new_item in [53, 55, 57, 59, 60, 62, 64, 65, 67, 69, 71, 72, 74, 76, 77, 79]:
                sequence.append(new_item)
        else:
            sequence.append(random.randint(low, high+1))
    
    return sequence

temp = random.sample(range(13), 6)
intervals = []
for x in temp:
    intervals.append(x)
    intervals.append(-x)
print(intervals)
intervals = [8, -8, 5, -5, 4, -4, 6, -6, 9, -9, 12, -12]
seq = generate_sequence(1000, 53, 79, intervals, random_state=4)
```
