```python
from music21 import *
s = converter.parse('tinyNotation: 2/4 ' + 'CC8 GG EE GG CC GG EE GG CC AA FF AA CC AA FF AA DD BB FF BB DD BB FF BB C1 '* 20)
s.show()
```

```python
mf = midi.translate.streamToMidiFile(s)
mf.open('alberti.mid', 'wb')
mf.write()
mf.close()
```
